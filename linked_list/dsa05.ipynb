# Circular linked list

### Circular singly linked lists

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
if __name__ == "__main__":
    first = Node(2)
    second = Node(3)
    last = Node(4)
    
    first.next = second
    second.next = last
    last.next = first
```

```python
#insertion in an empty list
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
def insertinempty(last,data):
    if last is not None:
        return last
    
    new_node = Node(data)
    last = new_node
    last.next = last
    return last

def print_list(last):
    if last is None:
        return
    
    head = last.next
    while True:
        print(head.data, end=" ")
        head = head.next
        if head == last.next:
            break
    print()
    
if __name__ == "__main__":
    last = None
    last = insertinempty(last, 1)
    
    print("list after insertion: ", end="")
    print_list(last)
    
```

```python
#at the bginning of the circular linked list
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
def at_beg(head,data):
    new_node = Node(data)
    if head.next is None:
        head.next = new_node
        new_node.next = head
        
    curr = head
    while curr.next != head:
        curr = curr.next
    curr.next = new_node
    new_node.next = head
    head = new_node
    return head

def print_list(head):
    if head is None:
        return 
    curr = head
    while True:
        print(curr.data,end="")
        curr = curr.next
        if curr == head:
            break
    print()
    
if __name__ == "__main__":
    head = Node(2)
    second = Node(3)
    third = Node(4)
    
    head.next = second
    second.next = third
    third.next = head
    
    print("before insertion: ",end="")
    print_list(head)
    
    head = at_beg(head,1)
    print("after insertion: ",end="")
    print_list(head)
    
    
        
    
    
```

```python
#at the end
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
def at_end(head,data):
    new_node = Node(data)
    if head is None:
        head.next = new_node
        new_node.next = head
    
    curr = head
    while curr.next != head:
        curr = curr.next
    curr.next = new_node
    new_node.next = head
    
    return head

def print_list(head):
    if head is None:
        return
    curr = head
    while True:
        print(curr.data,end="")
        curr = curr.next
        if curr == head:
            break
    print()
    
if __name__ == "__main__":
    head = Node(1)
    second = Node(2)
    third = Node(3)
    
    head.next = second
    second.next = third
    third.next = head
    
    print("before insertion: ",end="")
    print_list(head)
    
    head = at_end(head,4)
    print("after insertion: ",end="")
    print_list(head)
    
```

```python
# at specific point
class Node:
    def __init__(self,data):
        self.data= data
        self.next = None

def at_spec(head,pos,data):
    new_node = Node(data)
    if pos == 0:
        new_node.next = head
        head = new_node
        
    curr = head
    for i in range(1,pos-1):
        curr = curr.next
        if curr is None:
            return head
    new_node.next = curr.next
    curr.next = new_node
    if new_node.next is None:
        new_node.next = head
    
    return head

def print_list(head):
    if head is None:
        return 
    curr = head
    while True:
        print(curr.data,end = "")
        curr = curr.next
        if curr == head:
            break
    print()
    
if __name__ == "__main__":
    head = Node(1)
    second = Node(2)
    third = Node(4)
    
    head.next = second
    second.next = third
    third.next = head
    
    print("before insertion: ",end="")
    print_list(head)
    
    head = at_spec(head,3,3)
    print("after insertion: ",end="")
    print_list(head)
    

 
```
