# Circular linked list

### Circular singly linked lists

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
if __name__ == "__main__":
    first = Node(2)
    second = Node(3)
    last = Node(4)
    
    first.next = second
    second.next = last
    last.next = first
```

```python
#insertion in an empty list
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
def insertinempty(last,data):
    if last is not None:
        return last
    
    new_node = Node(data)
    last = new_node
    last.next = last
    return last

def print_list(last):
    if last is None:
        return
    
    head = last.next
    while True:
        print(head.data, end=" ")
        head = head.next
        if head == last.next:
            break
    print()
    
if __name__ == "__main__":
    last = None
    last = insertinempty(last, 1)
    
    print("list after insertion: ", end="")
    print_list(last)
    
```

```python
#at the bginning of the circular linked list
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
def at_beg(head,data):
    new_node = Node(data)
    if head.next is None:
        head.next = new_node
        new_node.next = head
        
    curr = head
    while curr.next != head:
        curr = curr.next
    curr.next = new_node
    new_node.next = head
    head = new_node
    return head

def print_list(head):
    if head is None:
        return 
    curr = head
    while True:
        print(curr.data,end="")
        curr = curr.next
        if curr == head:
            break
    print()
    
if __name__ == "__main__":
    head = Node(2)
    second = Node(3)
    third = Node(4)
    
    head.next = second
    second.next = third
    third.next = head
    
    print("before insertion: ",end="")
    print_list(head)
    
    head = at_beg(head,1)
    print("after insertion: ",end="")
    print_list(head)
    
    
        
    
    
```

```python
#at the end
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
def at_end(head,data):
    new_node = Node(data)
    if head is None:
        head.next = new_node
        new_node.next = head
    
    curr = head
    while curr.next != head:
        curr = curr.next
    curr.next = new_node
    new_node.next = head
    
    return head

def print_list(head):
    if head is None:
        return
    curr = head
    while True:
        print(curr.data,end="")
        curr = curr.next
        if curr == head:
            break
    print()
    
if __name__ == "__main__":
    head = Node(1)
    second = Node(2)
    third = Node(3)
    
    head.next = second
    second.next = third
    third.next = head
    
    print("before insertion: ",end="")
    print_list(head)
    
    head = at_end(head,4)
    print("after insertion: ",end="")
    print_list(head)
    
```

```python
# at specific point
class Node:
    def __init__(self,data):
        self.data= data
        self.next = None

def at_spec(head,pos,data):
    new_node = Node(data)
    if pos == 0:
        new_node.next = head
        head = new_node
        
    curr = head
    for i in range(1,pos-1):
        curr = curr.next
        if curr is None:
            return head
    new_node.next = curr.next
    curr.next = new_node
    if new_node.next is None:
        new_node.next = head
    
    return head

def print_list(head):
    if head is None:
        return 
    curr = head
    while True:
        print(curr.data,end = "")
        curr = curr.next
        if curr == head:
            break
    print()
    
if __name__ == "__main__":
    head = Node(1)
    second = Node(2)
    third = Node(4)
    
    head.next = second
    second.next = third
    third.next = head
    
    print("before insertion: ",end="")
    print_list(head)
    
    head = at_spec(head,3,3)
    print("after insertion: ",end="")
    print_list(head)
    

 
```

## Deletion

```python
# of first node

class Node:
    def __init__(self,data):
        self.data= data
        self.next = None
        
def del_first(head):
    if head is None:
        return None
    elif head.next == head:
        head = None
        
    curr = head
    while curr.next != head:
        curr = curr.next
    curr.next = head.next
    head = curr.next
    
    return head
    
def print_list(head):
    if head is None:
        return 
    curr = head
    while True:
        print(curr.data,end = "")
        curr = curr.next
        if curr == head:
            break
    print()
    

if __name__ == "__main__":
    head = Node(1)
    second = Node(2)
    third = Node(4)
    
    head.next = second
    second.next = third
    third.next = head
    
    print("before deletion: ",end="")
    print_list(head)
    
    head = del_first(head)
    print("after deletion: ",end="")
    print_list(head)
```

```python
# at the last node
class Node:
    def __init__(self,data):
        self.data= data
        self.next = None
        
def at_end(head):
    if head is None:
        return None
    elif head.next == head:
        return None
    current = head
    while current.next.next != head:
        current = current.next
        
    current.next = head
    
    return head

def print_list(head):
    if head is None:
        return 
    current = head
    while True:
        print(current.data,end="")
        current = current.next
        if current == head:
            break
    print()
    
if __name__ == "__main__":
    head = Node(1)
    second = Node(2)
    third = Node(3)
    
    head.next = second
    second.next = third
    third.next = head
    
    print("before deletion: ",end="")
    print_list(head)
    
    head = at_end(head)
    print("after deletion: ",end="")
    print_list(head)
            
            
```

```python
# at specific node
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

def at_spec(head,pos):
    if head is None:
        return None
    elif head.next == head:
        return None
    if pos == 1:
        current = head
        while current.next != head:
            current = current.next
        current.next = head.next
        head = current.next
        return head
    
    current = head
    for i in range(1,pos-1):
        current = current.next
        if current.next == head:
            return head
    
    current.next = current.next.next
    
    return head

def print_list(head):
    if head is None:
        return 
    current = head
    while True:
        print(current.data,end="")
        current = current.next
        if current == head:
            break
    print()
    
if __name__ == "__main__":
    head = Node(1)
    second = Node(2)
    third = Node(3)
    
    head.next = second
    second.next = third
    third.next = head
    
    print("before deletion: ",end="")
    print_list(head)
    
    head = at_spec(head,2)
    print("after deletion: ",end="")
    print_list(head)
    
    
```

## Searching in a circular list

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
def searching(head,key):
    if head is None:
        return None
    elif head.data == key:
        return True
    
    current = head
    while current.next != head:
        if current.data == key:
            return True
        else:
            current = current.next
            
    if current.data == key:
        return True
    
    return False

def print_list(head):
    if head is None:
        return 
    current = head
    while True:
        print(current.data,end=" ")
        current = current.next
        if current == head:
            break
        
    print()

if __name__ == "__main__":
    head = Node(1)
    second = Node(2)
    third = Node(3)
    
    head.next = second
    second.next = third
    third.next = head
    
    print("Original list: ",end="")
    print_list(head)
    key = 4
    found = searching(head,key)
    
    if found:
        print(f"Value {key} found in the list.")
    else:
        print(f"Value {key} not found in the list.")
            
    
    
```
